```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_npz(fname):
    with np.load(fname) as fh:
        return fh[fh.files[0]]
```

```python
ROOT = Path('/data/datasets/MRI/numpy')
split = 'train'

folder_a = ROOT/split/'A'
folder_b = ROOT/split/'B'
fnames_a = sorted(list(folder_a.glob('*npz')))
fnames_b = sorted(list(folder_b.glob('*npz')))
```

```python
idx = 6
fname_a, fname_b = fnames_a[idx], fnames_b[idx]
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
for fname, ax in zip([fname_a, fname_b], axes.flatten()):
    image = load_npz(fname)
    ax.set_title(f'({image.min():.1f}, {image.max():.1f})')
    ax.imshow(image)
```
